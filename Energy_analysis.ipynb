### Total domestic consumption

```python
total_file_path = 'Data/GOV UK/Subnational_total_final_consumption_2005_2021.xlsx'
excel_Data = pd.read_excel(total_file_path, None)  # None to load all sheets

# Let's see what sheets are available
excel_Data.keys()
```

```python

gas = []
electricity=[]
coal=[]
total=[]
manufactured_fuels=[]
petroleum=[]
bio=[]

for year in range(2005, 2022):  
    df_year = pd.read_excel(total_file_path, sheet_name=str(year),skiprows=5)
    gas.append(df_year[df_year[df_year.columns[1]]=='England']['Gas:\nDomestic'].loc[4])
    electricity.append(df_year[df_year[df_year.columns[1]]=='England']['Electricity:\nDomestic'].loc[4])
    coal.append(df_year[df_year[df_year.columns[1]]=='England']['Coal:\nDomestic'].loc[4])
    manufactured_fuels.append(df_year[df_year[df_year.columns[1]]=='England']['Manufactured\nfuels:\nDomestic\n[note 3]'].loc[4])
    petroleum.append(df_year[df_year[df_year.columns[1]]=='England']['Petroleum:\nDomestic'].loc[4])
    bio.append(df_year[df_year[df_year.columns[1]]=='England']['Bioenergy \nand wastes:\nDomestic\n'].loc[4])


energies_types=['gas','electricity','coal','manufactured fuels','petroleum','bio mass']

print(gas)
energies=[gas,electricity,coal,manufactured_fuels,petroleum,bio]
    

colors = sns.color_palette('pastel')[0:7]

#plt.figure(figsize=(16,8))
plt.stackplot([i for i in range(2005,2022)],energies, labels=energies_types, colors=colors)
plt.legend(loc='lower left')
plt.title('Energy domestic consumption of dwellings by type of energy, from 2005 to 2021')
plt.savefig("Output/energy-consumption-2005-2021.png",dpi=300,bbox_inches='tight')
plt.ylabel('Total final energy consumption (ktoe)')
plt.show()
```

```python
eng_eff_path = 'Eng-energy-efficiency-score.xlsx'
eng_eff=pd.read_excel(eng_eff_path,skiprows=6)
meters_by_area_path='elec_domestic_MSOA_stacked_2010-22.csv'
meters_by_area=pd.read_csv(meters_by_area_path,encoding='ISO-8859-1')

meters_by_area=meters_by_area[meters_by_area['Year']==2022]
meters_by_area.head()

columns=eng_eff.columns

dic_df={}
eng_eff.iloc[:, -1] = pd.to_numeric(eng_eff.iloc[:, -1], errors='coerce')
eng_eff.iloc[:, -2] = pd.to_numeric(eng_eff.iloc[:, -2], errors='coerce')
eng_eff.iloc[:, -3] = pd.to_numeric(eng_eff.iloc[:, -3], errors='coerce')
eng_eff.iloc[:, -4] = pd.to_numeric(eng_eff.iloc[:, -4], errors='coerce')
eng_eff.iloc[:, -5] = pd.to_numeric(eng_eff.iloc[:, -5], errors='coerce')

for i in eng_eff['Local authority district code'].unique()[0:5]:
    dic_df[f'{i}']={}
    dic_df[f'{i}']['total']=eng_eff[eng_eff['Local authority district code']==i][columns[-5]].sum()
    dic_df[f'{i}']['ex_house']=eng_eff[eng_eff['Local authority district code']==i][columns[-4]].sum()
    dic_df[f'{i}']['new_house']=eng_eff[eng_eff['Local authority district code']==i][columns[-3]].sum()
    dic_df[f'{i}']['ex_flat']=eng_eff[eng_eff['Local authority district code']==i][columns[-2]].sum()
    dic_df[f'{i}']['new_flat']=eng_eff[eng_eff['Local authority district code']==i][columns[-1]].sum()
    dic_df[f'{i}']['meters']=meters_by_area[meters_by_area['LA_code']==i]['Meters'].sum()
    
#     nb_meters=meters_by_area[meters_by_area['MSOA_code']==i]['Meters']
#     total_meters+=nb_meters
#     total_eng_eff+=eng_eff[eng_eff['Middle super output layer (MSOA) code']==i][columns[0]]/nb_meters
#     ex_house+=eng_eff[eng_eff['Middle super output layer (MSOA) code']==i][columns[1]]/nb_meters
#     new_house+=eng_eff[eng_eff['Middle super output layer (MSOA) code']==i][columns[2]]/nb_meters
#     ex_flat+=eng_eff[eng_eff['Middle super output layer (MSOA) code']==i][columns[3]]/nb_meters
#     new_flat+=eng_eff[eng_eff['Middle super output layer (MSOA) code']==i][columns[4]]/nb_meters
    
# plt.bar([total_eng_eff,ex_house,new_house,ex_flat,new_flat], ['All dwellings','Existing houses', 'New houses', 'Existing flats', 'New flats'])

# # Ajouter des titres pour les axes et le graphique
# plt.xlabel('Catégories')
# plt.ylabel('Valeurs')
# plt.title('National mean of the energy efficiency')
    
```
