```python
import pandas as pd

# Assuming your CSV file is named 'data.csv'
file_path = 'Model/DSM_V1.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the DataFrame
print(df)
```

### Derivation of the inflows and outflows based on stock and death rate

```python
death_rate=0.007 #arbitrary
df['Outflow']=death_rate*df['Stock']
df['Inflow']=df['Stock change']+df['Outflow']
display(df)
```

```python
# Plotting multiple columns
plt.figure(figsize=(10, 6))
plt.plot(df['Inflow'], label='Inflow')
plt.plot(df['Outflow'], label='Outflow')
plt.plot(df['Stock change'], label='Stock Change')
# Add more plots for additional columns if needed
plt.title('Plot of Multiple Columns')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.legend()
plt.grid(True)
plt.show()
```
